```python
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
df = sns.load_dataset('penguins')
df

```

```python
df.dropna(inplace = True)
```

# Feature Engineering

#One hot encoding transforming categorical data into numeric

```python
df.sex.unique()
```

```python
sex = pd.get_dummies(df['sex'], drop_first = True)
sex
```

```python
island = pd.get_dummies(df['island'], drop_first = True)
island
```

```python
df = pd.concat([df, sex, island], axis = 1)
df
```

```python
df = df.drop(df.columns[1:2], 1)
df
```

```python
df = df.drop(df.columns[5:6],1)
df
```

```python
df1 = df.rename(columns = {'Male': 'sex'})
df1
```

# Splitting the dataset into training and test data

```python
#seperate the target variable from the features/predictor
X = df.drop('species', axis = 1)
y = df['species']
```

```python
X,y
```

```python
from sklearn.preprocessing import LabelEncoder

la = LabelEncoder()

y = la.fit_transform(y)

y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
```

```python
print(X_train)
print(X_test)
print(y_train)
print(y_test)
```

```python
print('X_train', X_train.shape)
print('X_test', X_test.shape)
print('y_train', y_train.shape)
print('y_test', y_test.shape)
```

# Training Random Forest Classification On Training Set

```python
#Creating the model

from sklearn.ensemble import RandomForestClassifier

classifier = RandomForestClassifier(n_estimators = 7, criterion = 'entropy', random_state = 0) #n_estimator = number of decision
                                                                                               # trees
#you can keep changing the n_estimator(number of decision tree) and the criterion(entropy or gini) to get the accuracy scores.

classifier.fit(X_train, y_train)

y_pred = classifier.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

from sklearn.metrics import classification_report, accuracy_score
```

```python
df.species.unique()
```

```python
cn = confusion_matrix(y_test, y_pred)
print(cn)

#               Predicted values
#Actual values  0     1      2

#    0          48    0      0    Here the model predicted that there were only 48 Adelie species and actually there were
#                                 48, NO false predictions(hence 0= Chinstrap, 0 = Gentoo).

#    1          1     15     0   Here the model predicted that there were 15 Chinstrap species and actually there were
#                                15, false prediction = 1 Adelie.

#    2          1     0     36   Here the model predicted that there were 36 Gentoo species and in reality ,
                                 # there were 36, false preiction = 1 Adelie.
    
                                 #In all, there was only two false predictions.
```

```python
accuracy_score(y_test, y_pred) # 99% accuracy
```

```python
print(classification_report(y_test, y_pred))
```
