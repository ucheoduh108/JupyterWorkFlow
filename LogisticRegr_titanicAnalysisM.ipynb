# Train Data

```python
X = dff.drop('survived', axis =1) #independent variable
y = dff['survived'] #dependent variable
```

```python
from sklearn.model_selection import train_test_split #to split the data into training and testing subset
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1); X_train,X_test #splitting the data
```

```python
#Building the model

from sklearn.linear_model import LogisticRegression    

logmodel = LogisticRegression(max_iter = 250)           #create an instance of the logisticRegression model

logmodel.fit(X_train,y_train)                          # fit the model

predictions = logmodel.predict(X_test)                 #make predictions
```

```python
#To evaluate the performance of the model 

from sklearn.metrics import classification_report 
```

```python
classification_report(y_test,predictions) # precision of the model is - 0.74, 0.72, 0.73
```

```python
from sklearn.metrics import confusion_matrix #TO calculate the accuracy
```

```python
confusion_matrix(y_test,predictions)
```

```python
from sklearn.metrics import accuracy_score #calculating the accuracy
```
